```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img,0,255)
```

```python
plt.imshow(edges)
```

```python
med_value = np.median(img)
```

```python
med_value
```

```python
lower = int(max(0,0.7*med_value))
upper = int(min(255,1.3*med_value))
```

```python
edges = cv2.Canny(img,lower,upper+100)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
plt.imshow(blurred_img)
```

```python
edges = cv2.Canny(blurred_img,lower,upper+50)
plt.imshow(edges)
```
